```python
import pandas as pd
import numpy as np
from sklearn.preprocessing import LabelEncoder, StandardScaler
from sklearn.model_selection import train_test_split

df = pd.read_csv('../data/superstore_clean.csv')


df.columns = df.columns.str.strip().str.replace(' ', '_').str.lower().str.replace('-', '_')

print("Shape:", df.shape)
print("Columns:", df.columns.tolist())
```

```python
df['order_date'] = pd.to_datetime(df['order_date'])

df['order_year']    = df['order_date'].dt.year
df['order_month']   = df['order_date'].dt.month
df['order_quarter'] = df['order_date'].dt.quarter

print("Date features added!")
df[['order_date', 'order_year', 'order_month', 'order_quarter']].head()
```

```python
print("What we want to PREDICT (target):")
print(df['profit'].describe())

print("\nSample profit values:")
print(df['profit'].head(10).values)
```

```python

features = [
    'sales',
    'quantity', 
    'discount',
    'order_year',
    'order_month',
    'order_quarter'
]

X = df[features]
y = df['profit']

print("Features shape:", X.shape)
print("Target shape:", y.shape)
print("\nFeatures preview:")
X.head()
```

```python
le = LabelEncoder()

df['category_encoded']    = le.fit_transform(df['category'])
df['region_encoded']      = le.fit_transform(df['region'])
df['segment_encoded']     = le.fit_transform(df['segment'])
df['ship_mode_encoded']   = le.fit_transform(df['ship_mode'])

print("Encoding done!")
print(df[['category', 'category_encoded']].drop_duplicates())
```

```python
features = [
    'sales',
    'quantity',
    'discount',
    'order_year',
    'order_month',
    'order_quarter',
    'category_encoded',
    'region_encoded',
    'segment_encoded',
    'ship_mode_encoded'
]

X = df[features]
y = df['profit']

print("Updated features shape:", X.shape)
X.head()
```

```python
scaler = StandardScaler()
X_scaled = scaler.fit_transform(X)

print("Before scaling - sales mean:", round(X['sales'].mean(), 2))
print("After scaling  - sales mean:", round(X_scaled[:, 0].mean(), 4))
```

```python
X_train, X_test, y_train, y_test = train_test_split(
    X_scaled, y, 
    test_size=0.2, 
    random_state=42
)

print("Training rows:  ", X_train.shape[0])
print("Testing rows:   ", X_test.shape[0])
print("\n80% of data used to TRAIN the model")
print("20% of data used to TEST how good it is")
```
